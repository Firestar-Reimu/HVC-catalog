```python
v_dev_max, v_dev_min = [], []
v_dev_max_py, v_dev_min_py = [], []

for i in range(0, 361, 10):
    va, vi = calc_v_dev(i, 0, model='univ', v_dev=0.0)
    v_dev_max.append(va)
    v_dev_min.append(vi)
    va_py, vi_py = calc_v_dev_py(i, 0, model='univ', v_dev=0.0)
    v_dev_max_py.append(va_py)
    v_dev_min_py.append(vi_py)

v_dev_max = np.array(v_dev_max)
v_dev_min = np.array(v_dev_min)
v_dev_max_py = np.array(v_dev_max_py)
v_dev_min_py = np.array(v_dev_min_py)

print(v_dev_max_py - v_dev_max)
print(v_dev_min_py - v_dev_min)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(np.arange(0, 361, 10), v_dev_max_py - v_dev_max, label='max')
plt.plot(np.arange(0, 361, 10), v_dev_min_py - v_dev_min, label='min')
plt.xlabel('Angle (degrees)')
plt.ylabel('V_dev max')
plt.legend()
plt.show()
```
