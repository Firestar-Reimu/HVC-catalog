```python
import os
import glob
import numpy as np
from astropy.io import fits

# 获取所有FITS文件
fits_files = sorted(glob.glob('./*_N.fits'))

fits_files
```

```python
# 定义要修改的header内容
updates = {
    'CTYPE3': ('VRAD', 'Radio velocity (linear)'),
    'SPECSYS': ('LSRK', 'Reference frame of spectral coordinates'),
    'VELOSYS': (0.0, '[m/s] Velocity towards source'),
    'OBSFREQ': (1.42040575177e+9, '[Hz] Observer frequency'),
    'RESTFRQ': (1.42040575177e+9, '[Hz] Line rest frequency'),
    'BUNIT': ('K', 'Units of pixel data values'),
    'CTYPE1': ('RA---CAR', 'Right Ascension (CAR projection)'),
    'CTYPE2': ('DEC--CAR', 'Declination (CAR projection)'),
    'CDELT1': (-1/60, '[deg] Pixel size in RA direction'),
    'CDELT2': (1/60, '[deg] Pixel size in DEC direction'),
}

# 批量修改
for filename in fits_files:
    try:
        with fits.open(filename, mode='update') as hdul:
            header = hdul[0].header
            data = hdul[0].data

            # 更新header
            for key, (value, comment) in updates.items():
                header[key] = (value, comment)

            crpix2 = header['CRPIX2']
            crval2 = header['CRVAL2']
            print(crpix2, crval2)
            if crval2 != 0:
                crpix2 -= crval2 * 60
                crval2 = 0
                # crpix2 = np.floor(crpix2) + 0.5

            print("crpix2 after adjustment =", crpix2)
            print("crval2 after adjustment =", crval2)
            header['CRPIX2'] = crpix2
            header['CRVAL2'] = crval2

            hdul.flush()
            print(f"已更新: {filename}")

    except Exception as e:
        print(f"错误: {filename} - {e}")

print("处理完成!")
```
