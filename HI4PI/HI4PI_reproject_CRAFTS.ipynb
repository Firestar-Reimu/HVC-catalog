```python
from spectral_cube import SpectralCube as sc
```

```python
input_file = "./v_dev_30/HI4PI_RA60_80_DEC-13_2_-.fits"
crafts_file = "../CRAFTS/South/v_dev_30/Baseline_corrected/CRAFTS_RA60_80_DEC-13_2_-_corrected.fits"
output_file = "./v_dev_30/Reprojected/HI4PI_RA60_80_DEC-13_2_-_reprojected.fits"
```

```python
cube = sc.read(input_file)
crafts_cube = sc.read(crafts_file)
reprojected_cube = cube.reproject(crafts_cube.header, use_memmap=True, block_size="auto", parallel=True, dask_method="memmap")
reprojected_cube.write(output_file, overwrite=True)
```
