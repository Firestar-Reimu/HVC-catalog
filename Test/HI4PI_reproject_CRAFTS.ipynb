```python
from spectral_cube import SpectralCube as sc
```

```python
input_file = "./hi4pi_test2.fits"
crafts_file = "./test2.fits"
output_file = "./hi4pi_test2_reprojected.fits"

cube = sc.read(input_file)
crafts_cube = sc.read(crafts_file)
target_header = crafts_cube.wcs.celestial.to_header()

target_header['NAXIS'] = cube.ndim
target_header['NAXIS1'] = crafts_cube.shape[2]
target_header['NAXIS2'] = crafts_cube.shape[1]
target_header['NAXIS3'] = cube.shape[0]
target_header['CTYPE3'] = cube.wcs.wcs.ctype[2]
target_header['CDELT3'] = cube.wcs.wcs.cdelt[2]
target_header['CRPIX3'] = cube.wcs.wcs.crpix[2]
target_header['CRVAL3'] = cube.wcs.wcs.crval[2]
target_header['CUNIT3'] = "m/s"

reprojected_cube = cube.reproject(target_header, parallel=True)
reprojected_cube.write(output_file, overwrite=True)
```
