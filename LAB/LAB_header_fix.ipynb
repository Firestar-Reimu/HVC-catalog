```python
from astropy.io import fits

# 读取FITS文件
input_file = 'lab.fit'  # 替换为你的文件名
output_file = 'LAB_ALLSKY_GAL_CAR.fits'  # 输出文件名

# 打开FITS文件
hdul = fits.open(input_file, do_not_scale_image_data=True)

# 获取header
header = hdul[0].header

# 修改坐标参考点到图像中心
# 对于 721 像素: 中心在 (721+1)/2 = 361
# 对于 361 像素: 中心在 (361+1)/2 = 181

header['CRVAL1'] = 0.0  # 银经中心改为0度
header['CRPIX1'] = 361.0  # 参考像素改为图像中心

header['CRVAL2'] = 0.0  # 银纬中心改为0度
header['CRPIX2'] = 181.0  # 参考像素改为图像中心

# 修改速度为 VRAD-LSR
header['CTYPE3'] = 'VRAD'
header['CUNIT3'] = 'm/s'
header['CNAME3'] = 'LSRK radio velocity'
header['SPECSYS'] = 'LSRK'
header['SSYSOBS'] = 'LSRK'
header['SSYSSRC'] = 'LSRK'

# 保存修改后的文件
hdul.writeto(output_file, overwrite=True)

print(f"已成功修改 header 并保存到 {output_file}")

# 关闭文件
hdul.close()
```
