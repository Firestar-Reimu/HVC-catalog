```python
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.wcs import WCS
from matplotlib.colors import PowerNorm

%matplotlib widget
```

```python
file1 = "HI4PI_RA60_80_DEC-13_2_HVC_moment_1.fits"
file2 = "./Figures/HI4PI_RA60_80_DEC-13_2_-_moment1.fits"

hdul1 = fits.open(file1)
hdr1 = hdul1[0].header
wcs1 = WCS(hdr1)
data1 = hdul1[0].data

hdul2 = fits.open(file2)
hdr2 = hdul2[0].header
wcs2 = WCS(hdr2)
data2 = hdul2[0].data
```

```python
# 创建整个图形，左右排列
fig = plt.figure(figsize=(15, 6), constrained_layout=True)

# 第一个子图（左侧）
ax1 = fig.add_subplot(1, 2, 1, projection=wcs1)
im1 = ax1.imshow(
    data1,
    origin="lower",
    aspect="auto",
    norm=PowerNorm(gamma=1, vmin=-150, vmax=-80),
    cmap="viridis",
)
ax1.coords[0].set_major_formatter('d')
ax1.coords[1].set_major_formatter('d')
ax1.grid(ls="--")
ax1.set_xlabel("RA (deg)")
ax1.set_ylabel("Dec (deg)")
ax1.set_title("Moment Mask")

# 第二个子图（右侧）
ax2 = fig.add_subplot(1, 2, 2, projection=wcs2)
im2 = ax2.imshow(
    data2,
    origin="lower",
    aspect="auto",
    norm=PowerNorm(gamma=1, vmin=-150, vmax=-80),
    cmap="viridis",
)
ax2.coords[0].set_major_formatter('d')
ax2.coords[1].set_major_formatter('d')
ax2.grid(ls="--")
ax2.set_xlabel("RA (deg)")
ax2.set_ylabel("")
ax2.set_title("DBSCAN")

# 添加共享的颜色条
cbar = fig.colorbar(im1, ax=[ax1, ax2], orientation='vertical')
cbar.set_label(r"$V_{\mathrm{LSR}}$ (km/s)")

# 调整布局并显示
plt.savefig("./Figures/HI4PI_RA60_80_DEC-13_2_HVC_moment1_comparison.png", dpi=300)
plt.show()
```
