```python
import pandas as pd

input_file = 'dr11_v2.0_LRS_catalogue.csv'
df = pd.read_csv(input_file)

print(f"成功读取文件: {input_file}")
print(f"原始数据行数: {len(df)}")
```

```python
# 应用筛选条件
# 60 < ra < 135 且 -13 < dec < 2
filtered_df = df[
    (df['ra'] > 60) &
    (df['ra'] < 135) &
    (df['dec'] > -13) &
    (df['dec'] < 2)
]
print(f"筛选后数据行数: {len(filtered_df)}")
```

```python
# 获取subclass列的所有唯一值
unique_subclasses = filtered_df['subclass'].unique()

# 获取subclass列的值计数
subclass_counts = filtered_df['subclass'].value_counts()
print(f"\n'subclass' 列共有 {len(unique_subclasses)} 个不同的值:\n")

# 按字母顺序排序并输出所有可能的值
for subclass in sorted(unique_subclasses):
    count = subclass_counts[subclass]
    # 处理NaN值
    if pd.isna(subclass):
        print(f"  - [空值/NaN]: {count} 行")
    else:
        print(f"  - '{subclass}': {count} 行")
```

```python
# 筛选出subclass以O、B、A开头的行
oba_df = filtered_df[filtered_df['subclass'].str.startswith(('O', 'B', 'A'), na=False)]

# 按ra从小到大、dec从小到大排序
oba_df_sorted = oba_df.sort_values(by=['ra', 'dec'], ascending=[True, True])

# 写入新文件
oba_df_sorted.to_csv('LAMOST_DR11_RA60_135_DEC-13_2_OBA.csv', index=False)
print(f"数据已保存到: LAMOST_DR11_RA60_135_DEC-13_2_OBA.csv")
print(f"文件包含 {len(oba_df_sorted):,} 行，{len(oba_df_sorted.columns)} 列")
```
