```python
from astropy.io import fits

file = "./Taurus/Taurus_CRAFTS.fits"
hdul = fits.open(file)
hdr = hdul[0].header
data = hdul[0].data

if hdr['CRVAL2'] != 0:
    hdr['CRPIX2'] -= hdr['CRVAL2'] / hdr['CDELT2']
    hdr['CRVAL2'] = 0
hdr['BUNIT'] = 'K'
hdr['RESTFRQ'] = 1420405751.77
hdr['OBSFREQ'] = 1420405751.77
hdr['CTYPE1'] = 'RA---CAR'  # "RA---CAR"
hdr['CTYPE2'] = 'DEC--CAR'  # "DEC--CAR"
hdr['CTYPE3'] = 'VRAD'
hdr["SPECSYS"] = 'LSRK'
hdr["VELOSYS"] = 0.0
# del hdr["BLANK"]


fits.writeto(file, data, hdr, overwrite=True)
hdr
```
